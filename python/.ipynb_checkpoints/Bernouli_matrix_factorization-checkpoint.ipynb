## Implementing Algorithm

```python
from IPython.display import Image
Image(filename='algorithm.png', width=500) 
```

```python
def logit(u, v):
    return (1 / (1 + np.exp(-np.dot(u, v))))
```

```python
Image('matrix_data.png', width=300)
```

```python
data = np.array(
    [[0, 1, np.nan, 1, np.nan, np.nan], 
     [1, np.nan, 1, 0, 0, np.nan], 
     [np.nan, 1, 1, np.nan, np.nan, 0],
     [0, np.nan, 0, np.nan, 1, 1]
    ])
```

```python
Image(filename='user_random.png', width=300) 
```

```python
U = np.array([
    [0.99, 0.26, 0.55],
    [0.77, 0.77, 0.85],
    [0.20, 0.27, 0.35],
    [0.11, 0.96, 0.13],
], dtype=np.float64)

V = np.array([
    [0.91, 0.15, 0.27],
    [0.54, 0.54, 0.79],
    [0.31, 0.57, 0.09],
    [1.00, 0.83, 0.75],
    [0.68, 0.03, 0.05],
    [0.35, 0.50, 0.75],
], dtype=np.float64)
```

```python
Image(filename='item_random.png', width=300) 
```

```python
u_old = np.array([
    [0.99, 0.26, 0.55],
    [0.77, 0.77, 0.85],
    [0.20, 0.27, 0.35],
    [0.11, 0.96, 0.13],
], dtype=np.float16)

v_old = np.array([
    [0.91, 0.15, 0.27],
    [0.54, 0.54, 0.79],
    [0.31, 0.57, 0.09],
    [1.00, 0.83, 0.75],
    [0.68, 0.03, 0.05],
    [0.35, 0.50, 0.75],
], dtype=np.float16)
```

```python
m = 10
k = 3
temp = 0
# u = np.random.uniform(0, 1, size=(data.shape[0], k))
# v = np.random.uniform(0, 1, size=(k, data.shape[1]))
gamma = 0.1
eta = 0.01
for _ in range(m):
    for u in range(data.shape[0]):
        delta = np.zeros(k)
        for i in range(data.shape[1]):
            if not np.isnan(data[u, i]):
                for f in range(k):
                    if data[u, i] == 1:       
                        delta[f] += (1 - logit(U[u, :], V[i, :])) * V[i, f]
                    else:                       
                        delta[f] -= logit(U[u, :], V[i, :]) * V[i, f]
        for f in range(k):
            U[u, f] += gamma * (delta[f] - eta * U[u, f])
            
    for u in range(data.shape[1]):
        theta = np.zeros(k)
        for i in range(data.shape[0]):
            if not np.isnan(data[i, u]):
                for f in range(k):
                    if data[i, u] == 1:
                        theta[f] += (1 - logit(U[i, :], V[u, :])) * U[i, f]
                    else:
                        theta[f] -= logit(U[i, :], V[u, :]) * U[i, f]
        for f in range(k): 
            V[u, f] += gamma * (theta[f] - eta * V[u, f])
```
